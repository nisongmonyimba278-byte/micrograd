```python
{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# Topology optimisation of a microfluidic concentration gradient generator\n",
    "\n",
    "This notebook demonstrates the core functionality of `micrograd`:\n",
    "1. Automatic design of a microfluidic channel network that produces a linear concentration gradient.\n",
    "2. Visualisation of the final density field and outlet profile.\n",
    "3. Export of the optimised geometry as an STL file for fabrication.\n",
    "\n",
    "Run the cells below to reproduce the headline result of the manuscript."
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "from micrograd import GradientGeneratorOptimizer\n",
    "import matplotlib.pyplot as plt\n",
    "%matplotlib inline"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Create the optimiser with a linear target (0 → 1 across the outlet)\n",
    "opt = GradientGeneratorOptimizer(\n",
    "    target_expr=lambda x: x[1] / 500e-6,\n",
    "    Lx=2000e-6, Ly=500e-6,\n",
    "    nx=80, ny=20,\n",
    "    V_star=0.5\n",
    ")\n",
    "\n",
    "# Run the optimisation for 80 iterations (this may take ~20 min)\n",
    "rho_phys = opt.run(max_iter=80, beta_continuation=[1, 2, 4, 8, 16])"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Visualise the optimised density field\n",
    "opt.plot()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Plot the convergence history\n",
    "history = opt.history\n",
    "fig, ax = plt.subplots(figsize=(6,4))\n",
    "ax.plot(history[:,0], history[:,2], label='Total objective')\n",
    "ax.set_xlabel('Iteration')\n",
    "ax.set_ylabel('Objective J')\n",
    "ax.set_yscale('log')\n",
    "ax.legend()\n",
    "plt.show()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Export the optimised channel as an STL file (3D‑printable)\n",
    "opt.export_stl(\"optimised_gradient_generator.stl\", threshold=0.5)\n",
    "print(\"STL file saved.\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "The resulting STL file can be viewed in ParaView, imported into a CAD package, or sent to a 3D printer. The channel network achieves a linear concentration profile with ~50% lower hydraulic resistance than the classic Christmas tree design."
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3 (micrograd-env)",
   "language": "python",
   "name": "micrograd-env"
  },
  "language_info": {
   "name": "python",
   "version": "3.10.13"
  }
 },
 "nbformat": 4,
 "nbformat_minor": 5
}
```
